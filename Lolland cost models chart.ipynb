```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
font = {'size': 14}
matplotlib.rc('font', **font)
```

```python
df = {'bus Model 1': [24.66223437,16.92,12.596,10.011,8.46],
      'bus Model 3': [22.37873048,15.12062422,11.26872,8.945990728,7.63186],
      'DRT Model 1': [13.536,11.6372,10.528,10.4058,10.622],
      'DRT Model 2': [15.98,14.7533,14.1,13.54227549,12.878],
      'DRT Model 3': [11.418771532600001, 10.546653068600001, 10.2017201814, 9.8537148948, 9.4696261948]
}

df = pd.DataFrame(df)
```

```python
plt = df.plot(legend = True, style=['o-b','s-b','o-r','x-r','s-r'])
plt.set_xlim(-0.5,4.5)
plt.set_xticks([0,1,2,3,4])
plt.set_xticklabels(['50%', '75%', '100%', '125%', '150%'])
plt.set_xlabel('Demand level')
plt.set_ylabel('Cost per trip, EUR')
```

```python
fig = plt.get_figure()
fig.savefig('data/cost_chart.svg', bbox_inches='tight')
```
