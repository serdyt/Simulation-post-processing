```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
font = {'size': 14}
matplotlib.rc('font', **font)
markersize = {'markersize': 9}
matplotlib.rc('lines', **markersize)
```

```python
df = {'bus Model 1': [244.4890162,168.380829,124.8903122,99.83870968,84.62705706],
      'bus Model 3': [221.8515045,150.4394593,111.5829585,89.20066252,76.04131315],
      'DRT Model 1': [130.14441,117.4982446,110.0074943,110.3217748,105.0353269],
      'DRT Model 2': [185.0953856,166.9348823,159.7306886,152.0187901,145.3036136],
      'DRT Model 3': [129.1121469,117.6472445,112.2347759,109.1134223,104.8191182]
}

df = pd.DataFrame(df)
```

```python
df = df*0.094
```

```python
p = df.plot(legend = True, style=['o-b','s-b','+-r','s-r','x-r'])
p.set_xlim(-0.5,4.5)
p.set_ylim(0,35)
p.set_xticks([0,1,2,3,4])
p.set_xticklabels(['50%', '75%', '100%', '125%', '150%'])
p.set_xlabel('Demand level')
p.set_ylabel('Cost per trip, EUR')
p.grid(axis="y")
# plt.legend(bbox_to_anchor=(1.0, 1.13))
```

```python
fig = p.get_figure()
fig.savefig('data/cost_chart.svg', bbox_inches='tight')
```
